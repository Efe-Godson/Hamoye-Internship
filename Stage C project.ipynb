```python
#importing libraries 

import pandas as pd 
import requests
from bs4 import BeautifulSoup
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
```

```python
#assignig the url to a variable 
url = "https://www.visualcapitalist.com/best-performing-stocks-decade/"
    
#disguising request to website    
headers = {
    'User-Agent': 
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'
}

#using the request library to request from the webpage
response = requests.get(url, headers = headers )

#using Beautifulsoup to get the html markup that costructs the webpage
soup = BeautifulSoup(response.content,'html.parser')
```

```python
#finding the tables in the webpage
table= soup.find_all('table')
```

```python
#passing the table markup to pandas
df=pd.read_html(str(table[0]))
```

# DATA EXPLORATION AND CLEANING

```python
df[0].info()
```

```python
df2[0].info()
```

```python
#removing the currency sign and commas
df[0]['Final Value of $100 Investment'] = df[0]['Final Value of $100 Investment'].str.replace('$','').str.replace(',','')
```

```python
#changing the data type to integer
df[0]['Final Value of $100 Investment']= df[0]['Final Value of $100 Investment'].astype(int)
```

```python
df[0].sort_values(['Final Value of $100 Investment','Company'],ascending= False).head(5)
```
